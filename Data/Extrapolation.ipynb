```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
import parmap
import pickle
```

```python
df = pd.read_csv('kpro_data_train.csv')
print(df.shape)

with open(file='KPro_Train_entry.pickle', mode='rb') as f:
    entry = pickle.load(f)
with open(file='KPro_Train_smiles.pickle', mode='rb') as f:
    smiles = pickle.load(f)

tqdm.pandas()
df['3d'] = df['SMILES'].progress_apply(lambda x: smiles[x]['3d'] if x in smiles.keys() else 'NaN')
df = df[df['3d'] != 'NaN'].reset_index(drop = True)
df['pc'] = df['SMILES'].progress_apply(lambda x: smiles[x]['pc'] if x in smiles.keys() else 'NaN')
df = df[df['pc'] != 'NaN'].reset_index(drop = True)
df['mc'] = df['SMILES'].progress_apply(lambda x: smiles[x]['mc'] if x in smiles.keys() else 'NaN')
df = df[df['mc'] != 'NaN'].reset_index(drop = True)
df['mg'] = df['SMILES'].progress_apply(lambda x: smiles[x]['mg'] if x in smiles.keys() else 'NaN')
df = df[df['mg'] != 'NaN'].reset_index(drop = True)
df['msa'] = df['Name'].progress_apply(lambda x: entry[x]['msa'] if x in entry.keys() else 'NaN')
df = df[df['msa'] != 'NaN'].reset_index(drop = True)
df
```

```python
del df['3d']
del df['pc']
del df['mc']
del df['mg']
del df['msa']
df
```

```python
df.to_csv('kpro_data_train_desc.csv', index=False)
```

# Kinase Extrapolation

```python
train = pd.read_csv('kpro_data_train_desc.csv')
test = pd.read_csv('kpro_data_test_desc.csv')
```

```python
grp = pd.read_excel('./Dunbrack/si5.xlsx')[['5_Uni_acc', '13_Group_in_Uni']]
grp.columns = ['E', 'Group']
grp
```

```python
total = pd.read_csv('kpro_data_train_desc.csv')
total
```

```python
total = pd.merge(total, grp, left_on='Entry', right_on='E', how='inner')
del total['E']
total
```

```python
gs = []
ns = []
vs = []
vss = []
for g in total['Group'].unique():
    gdf = total[total['Group'] == g].reset_index(drop=True)
    gs.append(g)
    ns.append(gdf.shape[0])
    vs.append(np.var(gdf['Value'].tolist()))
    vss.append(gdf['Value'].tolist())
```

```python
g_df = pd.DataFrame({'G':gs, 'N':ns, 'V':vs, 'VS':vss})
g_df = g_df.sort_values(['N'], ascending=False)
g_df
```

```python
# TYR, CMGC, AGC
```

```python
gs = []
es = []
ms = []
ns = []
vs = []
vss = []
for g in ['TYR', 'CMGC', 'AGC']:
    gdf = total[total['Group'] == g].reset_index(drop=True)
    for e in gdf['Entry'].unique():
        edf = gdf[gdf['Entry'] == e].reset_index(drop=True)
        gs.append(g)
        es.append(e)
        ms.append(gdf[gdf['Entry'] == e]['Name'].iloc[0])
        ns.append(edf.shape[0])
        vs.append(np.var(edf['Value'].tolist()))
        vss.append(edf['Value'].tolist())
```

```python
e_df = pd.DataFrame({'G':gs, 'E':es, 'M':ms, 'N':ns, 'V':vs, 'VS':vss})
TYR = e_df[e_df['G'] == 'TYR'].sort_values(['N'], ascending=False).reset_index(drop=True)
CMGC = e_df[e_df['G'] == 'CMGC'].sort_values(['N'], ascending=False).reset_index(drop=True)
AGC = e_df[e_df['G'] == 'AGC'].sort_values(['N'], ascending=False).reset_index(drop=True)
TYR.shape, CMGC.shape, AGC.shape
```

```python
testtype.columns = ['G', 'Entry', 'M', 'Counts', 'Varation', 'VS']
del testtype['G']
del testtype['M']
del testtype['VS']
testtype
```

```python
test = total[total['Entry'].isin(testtype['Entry'])].reset_index(drop=True)
test = pd.merge(test, testtype, left_on='Entry', right_on='Entry', how='left')
test
```
